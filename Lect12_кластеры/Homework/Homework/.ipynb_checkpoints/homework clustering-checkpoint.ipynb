### Данные

В предложенных файлах информация с публичных слушаний Москвы по правилам землепользования и застройки (ПЗЗ). В них комментарии жителей города были застенагрофированы, проклассифицированы (за/против) и нанесены на карту. Данные предоставлены в 2 вариантах, для задания можно использовать любой:
* geo_comment.xlsx
    * **comment** - комментарий одного или списка жителей к проект
    * **multiplier** - количество авторов комментария (может быть 1, может быть список)
    * **x, y** - координаты адреса, по которому был дан определённой комментарий
    * **comment_class** - за (1) / против (-1)
* geo.xlsx - те же данные, но без текстов комментариев и по 1 голосу на строку (ранее в 1 строке могло быть **multiplier** > 1 голоса)
    * **x, y** - координаты адреса, по которому был дан определённой комментарий
    * **comment_class** - за (1) / против (-1)

* для тех, кто уже попробовал работу с текстом: выделить основные тематики комментариев жителей, можно использовать всю имеющуюся дополнительную информацию

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, AffinityPropagation
from sklearn.metrics import adjusted_rand_score, silhouette_score
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
#читаем файл эксель
msk = pd.read_excel('./geo_comment.xlsx')
msk.head()
```

```python
#строим точки по координатам из файла
plt.scatter(msk['x'], msk['y'], cmap='autumn', s=60)
```

```python
# сделаем дата фрейм из координат
geo = np.array([msk['x'], msk['y']])
geo = {'x': msk['x'], 'y': msk['y']}
```

```python
geo_df = pd.DataFrame(geo)
geo_df
```

```python
#Сделаем кластеризацию алгоритмом kmeans
clf_kmeans = KMeans(n_clusters=3)
clusters_kmeans = clf_kmeans.fit_predict(geo_df)
clusters_kmeans
```

```python
#отобразим кластеры на картинке
plt.scatter(msk['x'], msk['y'], c=clusters_kmeans, cmap='winter', s=60)
```

```python
k_inertia = []
ks = range(1,11)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(geo_df, )
    k_inertia.append(clf_kmeans.inertia_)
```

```python
plt.plot(ks, k_inertia)
```

```python
# по локтю видим, что оптимально 3 кластера
```

# аналогично делаем для точек с учетом голосования (проклассифицированы (за/против))

```python
#отобразим точки на карте с учетом голосования, т.е. параметра 'comment_class'
plt.scatter(msk['x'], msk['y'], c = msk['comment_class'], cmap='autumn', s=60)
```

```python
geo_class = {'x': msk['x'], 'y': msk['y'], 'comment_class': msk['comment_class']}
```

```python
geo_class_df = pd.DataFrame(geo_class)
geo_class_df
```

```python
#Сделаем кластеризацию алгоритмом kmeans
clf_kmeans = KMeans(n_clusters=2)
clusters_kmeans_class = clf_kmeans.fit_predict(geo_class_df)
clusters_kmeans_class 
```

```python
#отобразим кластеры на картинке
plt.scatter(msk['x'], msk['y'], c=clusters_kmeans_class , cmap='winter', s=60)
```

```python
# при n_clusters=2 получили тоже самое, что на картинке ранее
```

```python
k_inertia = []
ks = range(1,11)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(geo_class_df, )
    k_inertia.append(clf_kmeans.inertia_)
```

```python
plt.plot(ks, k_inertia)
```
